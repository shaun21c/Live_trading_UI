Set API parameters using Env params.

```python
AppKey = os.getenv('HTS_APP_KEY')
AppSecret = os.getenv('HTS_APP_SECRET')
```

```python
import requests

# API 엔드포인트 및 필요 정보 설정
url = "https://openapi.koreainvestment.com:9443/oauth2/tokenP"  # 실전 도메인
headers = {
    "Content-Type": "application/json; charset=UTF-8"
}
body = {
    "grant_type": "client_credentials",
    "appkey": AppKey,
    "appsecret": AppSecret
}

# POST 요청으로 접근토큰 발급
response = requests.post(url, headers=headers, json=body)

# 응답 처리
if response.status_code == 200:
    result = response.json()
    access_token = result.get("access_token")
    token_expiry = result.get("access_token_token_expired")
    print("발급된 접근 토큰:", access_token)
    print("토큰 만료 시간:", token_expiry)
else:
    print("토큰 발급 실패:", response.status_code, response.text)
```

## API 이용 Ticker 정보 GET

해외주식 현재가

```python
# API 엔드포인트 및 헤더 설정
url = "https://openapi.koreainvestment.com:9443/uapi/overseas-price/v1/quotations/price"
headers = {
    "Content-Type": "application/json; charset=utf-8",
    "authorization": f"Bearer {access_token}",
    "appkey": AppKey,
    "appsecret": AppSecret,
    "tr_id": "HHDFS00000300"  # 실전투자용 거래 ID
}

# 요청 파라미터 설정
params = {
    "AUTH": "",  # 사용자 권한정보, 기본값은 ""
    "EXCD": "NAS",  # 거래소 코드 (예: NAS = 나스닥)
    "SYMB": "TSLA"  # 종목 코드 (예: TSLA = 테슬라)
}

if __name__ == "__main__":
    while True:
        # GET 요청 보내기
        response = requests.get(url, headers=headers, params=params)

        # 응답 처리
        if response.status_code == 200:
            data = response.json()
            if data.get("rt_cd") == "0":
                output = data.get("output")
                print("현재가:", output.get("last"))
                print("거래량:", output.get("tvol"))
            else:
                print("API 응답 오류:", data.get("msg1"))
        else:
            print("HTTP 요청 오류:", response.status_code, response.text)
        
        # 일정 시간 대기 (예: 60초)
        time.sleep(0.1)
```
